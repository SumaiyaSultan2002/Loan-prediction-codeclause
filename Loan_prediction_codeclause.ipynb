```python
#import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
#import the dataset
df=pd.read_csv("/content/train loan prediction.csv")
```

# Data Preprocessing

```python
#number of null values in each column
df.isnull().sum()
```

```python
#Handle missing value 
df=df.drop("Loan_ID",axis=1)

#dropping columns with missing values 
columns=['Gender','Dependents','LoanAmount','Loan_Amount_Term']
df=df.dropna(subset=columns)

df.isnull().sum()
```

```python
#Handling the missing values in the columns "Self_Employed" and "Credit_History"
#Find the term that occurs most and replace the missing value with it

#Most occuring term in the columns "Self_Employed"
df['Self_Employed'].mode()[0]
```

```python
#replacing the missing value with mode
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
#Most occuring term in the columns "Credit_History"
df['Credit_History'].mode()[0]
```

```python
#replacing the missing value with mode
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df.isnull().sum()
```

Handling categorical columns

```python
df.head()
```

```python
df['Dependents']=df['Dependents'].replace(to_replace="3+",value='4')
```

```python
df['Loan_Status'].unique()
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0}).astype('int')
df['Married'] = df['Married'].map({'Yes':1,'No':0}).astype('int')
```

```python
df['Property_Area'] = df['Property_Area'].map({'Rural':0,'Semiurban':2,'Urban':1}).astype('int')
df['Loan_Status'] = df['Loan_Status'].map({'Y':1,'N':0}).astype('int')
```

```python
df['Education'] = df['Education'].map({'Graduate':1,'Not Graduate':0}).astype('Int64')
```

```python
df['Self_Employed'] = df['Self_Employed'].map({'Yes':1,'No':0}).astype('Int64')
```

```python
df.head()
```

# Feature Scaling


```python
X = df.drop('Loan_Status',axis=1)
y = df['Loan_Status']
df.sample(5)
```

```python
cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
```

```python
st = StandardScaler()
X[cols]=st.fit_transform(X[cols])
X
```

# Split the Dataset

```python
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

# Training and testing different ML Model


```python
#DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_valid)
print("Training Accuracy :", model.score(x_train, y_train))
print("Validation Accuracy :", model.score(x_valid, y_valid))
# calculating the f1 score for the validation set
print("f1 score :", f1_score(y_valid, y_pred))
# confusion matrix
cm = confusion_matrix(y_valid, y_pred)
print(cm)
```

```python
#RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_valid)
print("Training Accuracy :", model.score(x_train, y_train))
print("Validation Accuracy :", model.score(x_valid, y_valid))

# calculating the f1 score for the validation set
print("F1 score :", f1_score(y_valid, y_pred))

# confusion matrix
cm = confusion_matrix(y_valid, y_pred)
print(cm)
```

```python
#SVC
model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_valid)
print("Training Accuracy :", model.score(x_train, y_train))
print("Validation Accuracy :", model.score(x_valid, y_valid))
# calculating the f1 score for the validation set
print("f1 score :", f1_score(y_valid, y_pred))
# confusion matrix
cm = confusion_matrix(y_valid, y_pred)
print(cm)
```

```python
#Logistic Regression
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_valid)
print("Training Accuracy :", model.score(x_train, y_train))
print("Validation Accuracy :", model.score(x_valid, y_valid))
# calculating the f1 score for the validation set
print("f1 score :", f1_score(y_valid, y_pred))
# confusion matrix
cm = confusion_matrix(y_valid, y_pred)
print(cm)
```

# Save the model

```python
filename = 'finalized_model.sav'
joblib.dump(model, filename)
 
 
# load the model from disk
loaded_model = joblib.load(filename)
result = loaded_model.score(x_valid, y_valid)
print(result)
```
